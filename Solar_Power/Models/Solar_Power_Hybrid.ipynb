```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint('best_model.h5', monitor='val_rmse', save_best_only=True, mode='min', verbose=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate, Conv1D, Flatten, MultiHeadAttention
from tensorflow.keras.models import Model
```

```python
class DLinearModel(tf.keras.Model):
    def __init__(self, linear_units, dnn_units):
        super(DLinearModel, self).__init__()
        self.linear_layer = Dense(units=linear_units, activation=None)
        self.dnn_layers = [Dense(units, activation='relu') for units in dnn_units]
        self.final_layer = Dense(units=1, activation='linear')

    def call(self, inputs):
        linear_output = self.linear_layer(inputs)
        dnn_input = tf.concat([inputs, linear_output], axis=-1)
        dnn_output = dnn_input
        for layer in self.dnn_layers:
            dnn_output = layer(dnn_output)
        output = self.final_layer(dnn_output)
        return output
```

```python
class TCMixModel(tf.keras.Model):
    def __init__(self, input_size, hidden_sizes, kernel_sizes, num_layers, output_size):
        super(TCMixModel, self).__init__()

        self.temporal_layers = []
        for i in range(num_layers):
            self.temporal_layers.append(Conv1D(
                filters=hidden_sizes[i],
                kernel_size=kernel_sizes,
                padding='same',
                activation='relu'
            ))

        self.token_mixer = MultiHeadAttention(num_heads=8, key_dim=hidden_sizes[-1])
        self.fc = Dense(output_size)

    def call(self, x):
        for conv in self.temporal_layers:
            x = conv(x)
        x = self.token_mixer(x, x)
        x = Flatten()(x)
        x = self.fc(x)

        return x
```

```python
linear_units = 1
dnn_units = [64, 64, 32]
dlinear_model = DLinearModel(linear_units, dnn_units)
dlinear_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

input_size = x_train.shape[1]
hidden_sizes = [32, 64, 128]
kernel_sizes = 1
num_layers = 3
output_size = 1
tcmix_model = TCMixModel(input_size, hidden_sizes, kernel_sizes, num_layers, output_size)
tcmix_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

```

```python
def build_hybrid_model(dlinear_model, tcmix_model, input_shape):
    dlinear_input = Input(shape=(input_shape,))
    dlinear_output = dlinear_model(dlinear_input)

    tcmix_input = Input(shape=(input_shape,))
    tcmix_input_reshaped = tf.expand_dims(tcmix_input, axis=-1)
    tcmix_output = tcmix_model(tcmix_input_reshaped)

    combined_output = Concatenate()([dlinear_output, tcmix_output])
    final_output = Dense(1)(combined_output)

    hybrid_model = Model(inputs=[dlinear_input, tcmix_input], outputs=final_output)
    hybrid_model.compile(optimizer='adam', loss='mse', metrics=[rmse,'mae'])
    return hybrid_model

hybrid_model = build_hybrid_model(dlinear_model, tcmix_model, input_shape=x_train.shape[1])
```

```python
lambdas = [2.5468741654272455, 1.8680814672464994, 0.5464690175831447]
months = ['Jan', 'Feb', 'March', 'April', 'May', 'June',
          'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
results = []
```

```python
from scipy.stats import boxcox

def transform_data(df, lambdas):
    transformed_df = df.copy()
    for i, column in enumerate(df.columns[:-1]):
        data = df[df[column] > 0][column]
        transformed_data = boxcox(data, lambdas[i])
        transformed_df.loc[df[column] > 0, column] = transformed_data
    return transformed_df
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
from tensorflow.keras.utils import custom_object_scope
def load_hybrid_model(model_path):
    with custom_object_scope({'DLinearModel': DLinearModel, 'TCMixModel': TCMixModel, 'rmse': rmse}):
        return tf.keras.models.load_model(model_path)

model_path = '/content/best_model.h5'
model = load_hybrid_model(model_path)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    non_zero_indices = y_true != 0
    return np.mean(np.abs((y_true[non_zero_indices] - y_pred[non_zero_indices]) / y_true[non_zero_indices])) * 100

results = []

for month in months:
    data = pd.read_csv(f'{month}.csv')
    data.drop(['YEAR', 'Month', 'Day', 'Hour', 'Pressure', 'Precipitable Water', 'Wind Direction', 'Wind Speed'], axis=1, inplace=True)

    transformed_data = transform_data(data, lambdas)

    X_month = transformed_data.iloc[:, :-1].values
    y_month = transformed_data.iloc[:, -1].values

    X_month = scaler.fit_transform(X_month)

    X_month_reshaped = X_month.reshape((X_month.shape[0], X_month.shape[1], 1))

    y_pred = model.predict([X_month, X_month_reshaped]).flatten()

    rmse_val = rmse(y_month, y_pred)
    mae_val = mean_absolute_error(y_month, y_pred)
    mse_val = mean_squared_error(y_month, y_pred)
    mape_val = mape(y_month, y_pred)

    results.append([month, rmse_val, mae_val, mse_val, mape_val])

results_df = pd.DataFrame(results, columns=['Month', 'RMSE', 'MAE', 'MSE', 'MAPE'])

print(results_df)
```

```python
average_rmse = results_df['RMSE'].mean()
average_mae = results_df['MAE'].mean()
average_mse = results_df['MSE'].mean()
average_mape = results_df['MAPE'].mean()

annual_row = pd.DataFrame([['Annual', average_rmse, average_mae, average_mse, average_mape]], columns=['Month', 'RMSE', 'MAE', 'MSE', 'MAPE'])
results_df = pd.concat([results_df, annual_row], ignore_index=True)
```
